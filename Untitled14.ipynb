```python
import os
import json
import xml.etree.ElementTree as ET

# Function to parse XML files and extract annotation information
def parse_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    filename = root.find('filename').text
    width = int(root.find('size/width').text)
    height = int(root.find('size/height').text)

    annotations = []
    for obj in root.findall('object'):
        class_name = obj.find('name').text
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)

        annotations.append({
            'category_id': class_name,
            'bbox': [xmin, ymin, xmax - xmin, ymax - ymin],  # COCO bbox format: [x, y, width, height]
            'area': (xmax - xmin) * (ymax - ymin),
            'iscrowd': 0  # 0 for regular bounding box
        })

    return {
        'filename': filename,
        'width': width,
        'height': height,
        'annotations': annotations
    }

# Function to convert XML annotations to COCO format
def convert_to_coco(xml_file):
    # Convert single XML file to COCO format
    coco_data = parse_xml(xml_file)

    # Assuming you have predefined categories
    categories = ['Gun']

    # Initialize counters
    image_id = 0
    annotation_id = 0

    # Map category names to IDs
    category_id_map = {category_name: i for i, category_name in enumerate(categories)}

    # Update category IDs in annotations
    for annotation in coco_data['annotations']:
        annotation['category_id'] = category_id_map.get(annotation['category_id'])

    # Add categories to COCO data
    coco_data['categories'] = [{'id': i, 'name': category_name} for i, category_name in enumerate(categories)]

    return coco_data

# Directory containing XML annotation files
xml_dir = '/content/drive/MyDrive/WEAPON2/Labels'

# Output directory for COCO annotations
output_dir = '/content/drive/MyDrive/WEAPON2/OUTPUT'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Iterate over XML files in the input directory
for xml_file in os.listdir(xml_dir):
    if xml_file.endswith('.xml'):
        xml_path = os.path.join(xml_dir, xml_file)
        output_file = os.path.join(output_dir, f'{os.path.splitext(xml_file)[0]}.json')

        # Convert XML annotation to COCO format
        coco_data = convert_to_coco(xml_path)

        # Write COCO data to JSON file in the output directory
        with open(output_file, 'w') as f:
            json.dump(coco_data, f)

        print(f'Conversion complete for {xml_file}. COCO annotations saved to {output_file}')
```

```python
import os
import json
import tensorflow as tf
import numpy as np

# Step 1: Data Preparation
data_dir = '/content/drive/MyDrive/WEAPON2/OUTPUT'
gun_annotations = []

for filename in os.listdir(data_dir):
    if filename.endswith('.json'):
        with open(os.path.join(data_dir, filename), 'r') as file:
            json_data = json.load(file)
            for annotation in json_data['annotations']:
                if annotation['category_id'] == 'Gun':  # Adjust category ID if needed
                    gun_annotations.append({
                        'filename': annotation['filename'],
                        'bbox': annotation['bbox']
                    })

X_train = []
y_train = []

for annotation in gun_annotations:
    X_train.append(annotation['bbox'])
    y_train.append(1)  # Gun class

X_train = np.array(X_train)
y_train = np.array(y_train)

# Step 2: Model Training
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10)

# Step 3: Model Evaluation (optional)
# Evaluate the model on a separate test set

# Step 4: Model Deployment
# Save the model for inference
model.save('gun_detection_model.h5')
```
